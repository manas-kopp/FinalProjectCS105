```python
import pandas as pd
import matplotlib.pyplot as plt


df = pd.read_csv("drugLibTrain_raw.csv")

Douma = df['urlDrugName'].value_counts().head(12)

plt.figure(figsize=(10,5))
Douma.plot(kind='bar', color = 'red')

plt.title('Most reviewed drugs')
plt.xlabel('Drug name')
plt.ylabel('Number of reviews')
plt.xticks(rotation=45, ha='right')
plt.grid(axis = 'y', linestyle = '-', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("drugLibTrain_raw.csv")


scores = df['rating'].dropna().astype(int) 

plt.figure(figsize=(10,5))
plt.hist(scores, bins=range(1, 10), edgecolor = 'black')
plt.title('distribution of 1-10 effectiveness ratings')
plt.xlabel('Rating')
plt.ylabel('Number of reviews')
plt.xticks(range(1,10))
plt.grid(true)
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

df = pd.read_csv("drugLibTrain_raw.csv")

conditions = df['condition'].value_counts().head(10)

plt.figure(figsize=(10,6))
conditions.plot(kind='bar')
plt.title('Most common conditions during reviews')
plt.xlabel('condition')
plt.ylabel('number of reviews')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv('drugLibTrain_raw_sentiment (1).csv')
df['rating'] = pd.to_numeric(df['rating'])
df['reviewSentiment'] = pd.to_numeric(df['reviewSentiment'])
cleaner = df.dropna(subset=['rating', 'reviewSentiment'])
plt.figure(figsize=(10, 5))
sns.regplot(data=cleaner, x='rating', y='reviewSentiment', scatter_kws={'alpha':0.4}, line_kws={'color':'blue'})
plt.title('correlation between effectiveness score/rating and sentiment score of reviews')
plt.xtitle('rating')
plt.ylabel('Sentiment score')
plt.grid(True)
plt.tight_layout()
plt.show()


```
